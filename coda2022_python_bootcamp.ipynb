## 4ª meta: cruzamento de dados com outro dataset

```python
# rpa é Região Político Administrativa
rpa_url = 'http://dados.recife.pe.gov.br/dataset/c1f100f0-f56f-4dd4-9dcc-1aa4da28798a/resource/698d8fe1-d30a-485f-8d5d-307d46163d0c/download/bairro.csv'
```

```python
rpa = pd.read_csv(rpa_url, sep=';')
rpa
```

Criar uma nova coluna rpa_nome
1. Centro
2. Norte
3. Noroeste
4. Oeste
5. Sudoeste
6. Sul

Mapeamento de chaves para um valores também são conhecidos como **dicionários**.

```python
# mapeia os códigos numéricos para seus respectivos significados por extenso
rpa_de_para = {
    1: 'Centro',
    2: 'Norte',
    3: 'Noroeste',
    4: 'Oeste',
    5: 'Sudoeste',
    6: 'Sul',
}
# substitui os valores na coluna de acordo com o mapeamento acima 
coluna_rpa_traduzida = rpa['rpa'].replace(rpa_de_para)
coluna_rpa_traduzida
```

Apenas uma observação: o dataframe `rpa` não foi modificado pois a maioria dos métodos de `pandas` não modificam o objetos original e retornam uma **cópia**. Podemos ver que `rpa` permanece sem modificações.

```python
rpa
```

```python
# insere uma nova coluna em rpa
rpa['rpa_traduzido'] = coluna_rpa_traduzida
rpa
```

Quando lidamos com coleções, podemos realizar ações em cada elemento daquela coleção individualmente. Uma técnica muito utilizada para isso são os **laços** (ou iteração).

```python
# para cada elemento na lista
for bairro in acidentes_2021['bairro'].unique():
  # imprime o elemento
  print(bairro)
```

Também é possível fazer **condicionais** no código.

```python
# para cada elemento na lista
for bairro in acidentes_2021['bairro'].unique():
  # se o elemento não estiver na lista de bairros de rpa
  if bairro not in rpa['Bairro'].unique():
    # imprime o elemento
    print(bairro)
```

Também podemos adicionar elementos em uma lista por meio de **inserção em listas**.

```python
bairros_acidentes_com_erro = []
for bairro in acidentes_2021['bairro'].unique():
  if bairro not in rpa['Bairro'].unique():
    bairros_acidentes_com_erro.append(bairro)

bairros_acidentes_com_erro
```

E a lista pode ser **convertida em uma série** facilmente.

```python
bairros_acidentes_com_erro = []
for bairro in acidentes_2021['bairro'].unique():
  if bairro not in rpa['Bairro'].unique():
    bairros_acidentes_com_erro.append(bairro)

pd.Series(bairros_acidentes_com_erro).unique()
```

```python
# converte os valores da série para caixa alta
rpa['Bairro'].str.upper().unique()
```

```python
bairros_acidentes_com_erro = []
for bairro in acidentes_2021['bairro'].unique():
  if bairro not in rpa['Bairro'].str.upper().unique():
    bairros_acidentes_com_erro.append(bairro)

pd.Series(bairros_acidentes_com_erro).unique()
```

Alguns trechos de código podem ser reutilizados e se tornam mais legíveis se utilizarmos **funções**.

```python
# elimina os acentos dos valores de uma série
def magia(serie):
  return serie.str.normalize("NFKD").str.encode("ascii", "ignore").str.decode("utf8")

# chama a função com a série como argumento
magia(acidentes_2021['bairro'])
```

Mas a **magia** tem nome!

```python
def remove_acentos(serie):
  return serie.str.normalize("NFKD").str.encode("ascii", "ignore").str.decode("utf8")

remove_acentos(acidentes_2021['bairro'])
```

```python
bairros_acidentes_com_erro = []
for bairro in remove_acentos(acidentes_2021['bairro']).unique():
  if bairro not in rpa['Bairro'].str.upper().unique():
    bairros_acidentes_com_erro.append(bairro)

pd.Series(bairros_acidentes_com_erro).unique()
```

```python
exemplo_de_para = {
    'MADALENA': 'Ô MADALENA',
    'VÁRZEA': 'O MELHOR CANTINHO DA CIDADE',
    'VASCO DA GAMA': 'GIGANTE',
    'BOA VIAGEM': 'SEM COMENTÁRIOS',
}
acidentes_2021['bairro'].replace(exemplo_de_para)
```

```python
bairros_de_para = {
    'BAIRRO DO RECIFE': 'RECIFE',
    'JOANA BEZERRA': 'ILHA JOANA BEZERRA',
    'ALTO SANTA TERESINHA': 'ALTO SANTA TEREZINHA',
    'POCO DA PANELA': 'POCO',
}
bairros_acidentes_com_erro = []
bairros_acidentes_limpo = remove_acentos(acidentes_2021['bairro']).replace(bairros_de_para)
for bairro in bairros_acidentes_limpo.unique():
  if bairro not in rpa['Bairro'].str.upper().unique():
    bairros_acidentes_com_erro.append(bairro)

bairros_acidentes_com_erro = pd.Series(bairros_acidentes_com_erro)
bairros_acidentes_com_erro.unique()
```

```python
#adiciona nova coluna
acidentes_2021['bairro_limpo'] = bairros_acidentes_limpo
acidentes_2021
```

```python
#adiciona nova coluna
rpa['bairro_limpo'] = rpa['Bairro'].str.upper()
rpa
```

```python
acidentes_2021_rpa = acidentes_2021.merge(rpa, on='bairro_limpo')
acidentes_2021_rpa
```

```python
for ano in urls_anos:
  print(ano)
  print(urls_anos[ano])
```

```python
for ano, url in urls_anos.items():
  print(ano)
  print(url)
```

E com métodos como `.concat()` e `.merge()` podemos juntar os dataframes.

```python
acidentes = pd.DataFrame()
for ano, url in urls_anos.items():
  dataset_ano = pd.read_csv(url, sep=';')
  dataset_ano['ano'] = ano
  acidentes = pd.concat([acidentes, dataset_ano])

acidentes
```

```python
acidentes['bairro_limpo'] = remove_acentos(acidentes['bairro']).replace(bairros_de_para)
acidentes_rpa = acidentes.merge(rpa, on='bairro_limpo')
acidentes_rpa.info()
```

```python
acidentes_rpa = acidentes.merge(rpa, how='left', on='bairro_limpo')
acidentes_rpa.info()
```

Descartar linhas com valores nulos é simples e pode facilitar nossa análise.

```python
acidentes_regiao = acidentes_rpa.dropna(subset=['rpa_traduzido'])
acidentes_regiao.info()
```

```python
acidentes_regiao.groupby(['rpa_traduzido', 'ano']).size()
```
